```python
# There are 466 pges of it 
# so we will insert this into a loop
```

```python
df=pd.DataFrame() 
```

```python
df
```

```python

for i in range(1, 466):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8cb1d8a557776f6206aa9fd3a21be4b8&language=en-US&page={}'.format(i))
    
    # Create a DataFrame from the JSON response
    masoom = pd.DataFrame(response.json()['results'])[['id', 'title', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    
    # Append the new DataFrame to the existing DataFrame using concat
    df = pd.concat([df, masoom], ignore_index=True)
```

```python
df.to_csv('toprated_movies.csv')
```
